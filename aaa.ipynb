```python
# 1. 라이브러리를 이용하여 윈도우 위젯을 생성하고
# 2. 내부의 구역을 적절하게 나누고(프레임?, 라벨?)
# 3. 각 구역에 적절한 기능을 삽입하고
#    (버튼, 리스트박스, 라디오, 체크버튼, 텍스트박스, 캔버스)
# 4. 각 버튼을 함수와 연결시켜 이용한다.

# 사용할 라이브러리들을 import한다.
from tkinter import *
from tkinter.messagebox import showinfo

# 사용할 함수들 정리.
def tog():
    """배경색상 변경 버튼과 연동"""
    if window.cget('bg') =='yellow':
        window.configure(bg='gray')
    else:
        window.configure(bg='yellow')

def dialog_listbox():
    """listbox 선택 버튼과 연동"""
    try:
        return showinfo('선택', '당신의 선택 : ' + listbox.get(listbox.curselection()))
    except:
        pass

def dialog_radio():
    """radio 선택 버튼과 연동"""
    try:
        return showinfo('선택', '당신의 선택 : ' + book.get())
    except:
        pass


def dialog_checkbox():
    """checkbox 버튼과 연동"""
    str = '니 선택'
    if var_1.get() == 1: 
        str += '\nHTML5 in easy steps'
    if var_2.get() == 1: 
        str += '\nCSS3 in easy steps'
    if var_3.get() == 1: 
        str += '\nJavaScript in easy steps'
    showinfo('선택', str)

# 기본창
window = Tk()
window.title('싹 다 모아버려!')

label = Label(window, text="문구") # 라벨(?) 생성
label.pack(padx=50, pady=10) # 바깥여백 설정

btn_tog = Button(window, text='색상변경', command=tog) # 버튼 생성 및 함수 연결
btn_tog.pack(padx=50, pady=10) # 바깥여백 설정

frame = Frame(window) # 프레임(?) 생성
frame.pack(padx=50) # 바깥여백 설정

listbox = Listbox(frame) # 리스트박스 프레임 내에 생성
listbox.pack(side = RIGHT) # 정렬
listbox.insert(1, 'HTML5 in easy steps') # 리스트박스 내부 요소 추가
listbox.insert(2, 'CSS3 in easy steps')
listbox.insert(3, 'JavaScript in easy steps')

btn = Button(frame, text = 'Choose', command = dialog_listbox) # 리스트박스 버튼 연동
btn.pack(side=RIGHT, padx=5) # 버튼 설정

frame2 = Frame(window) # 프레임(?) 추가
frame2.pack(padx=30, pady=10) # 및 설정

book = StringVar() # radio버튼에서 전달된 문자열 자료를 보관할 객체 생성
# radio 생성 및 book 연결
radio_1 = Radiobutton(frame2, text = 'HTML5', variable = book, value = 'HTML5')
radio_2 = Radiobutton(frame2, text = 'CSS3', variable = book, value = 'CSS3')
radio_3 = Radiobutton(frame2, text = 'JS', variable = book, value = 'JavaScript')
# 각 radio 설정
radio_1.pack(side = LEFT)
radio_2.pack(side = LEFT)
radio_3.pack(side = LEFT)
# radio_1 체크해놓기
radio_1.select()

btn = Button(frame2, text = 'Choose', command = dialog_radio) # 버튼 생성 및 radio 연동
btn.pack(padx=5) # 버튼 설정

frame3 = Frame(window) # 프레임 생성

# 체크버튼의 데이터를 저장할 객체 생성
var_1 = IntVar()
var_3 = IntVar()
var_2 = IntVar()
# 체크버튼 생성 및 데이터 저장객체와 연동
book_1 = Checkbutton(frame3, text = 'HTML5', variable = var_1, onvalue = 1, offvalue = 0)
book_2 = Checkbutton(frame3, text = 'CSS3', variable = var_2, onvalue = 1, offvalue = 0)
book_3 = Checkbutton(frame3, text = 'JS', variable = var_3, onvalue = 1, offvalue = 0)
# 각 버튼 설정
book_1.pack(side = LEFT)
book_2.pack(side = LEFT)
book_3.pack(side = LEFT)

# 버튼 생성, 함수와 연동 및 설정
btn = Button(frame3, text = '선택', command = dialog_checkbox)
btn.pack(side = RIGHT, padx = 5)

img = PhotoImage(file = 'girl.gif') # 이미지 로드
small_img = PhotoImage.subsample(img, x = 2, y = 2) # 이미지 절반사이즈로 편집

# 라벨 생성 및 설정, 이미지 삽입
label = Label(window, image = img, bg = 'yellow')
label.pack(side = TOP)

# 버튼 생성 및 설정, 이미지 삽입
btn = Button(window, image = small_img)
btn.pack(side = LEFT, padx = 10)

txt = Text(window, width = 25, height = 7) # 텍스트 박스 생성
txt.image_create('1.0', image = small_img) # 내부 이미지 삽입
txt.insert('1.1', 'Python Fun!') # 문장 삽입
txt.pack(side = LEFT) # 텍스트박스 설정

can = Canvas(window, width = 100, height = 100, bg = 'green') # 캔버스 생성
can.pack(side = LEFT, padx = 10) # 설정
can.create_image((50, 50), image = small_img) # 이미지 삽입
can.create_line(30, 20, 57, 20, width = 5, fill = 'black') # 그리기

# 종료될 때까지 실행.
window.mainloop()
```

```python
from tkinter import *
import tkinter.messagebox as box
window=Tk()
window.title('Massage Box Example')
def dialog():
    var=box.askyesno('Message Box', 'Proceed?')
    if var==1:
        box.showinfo('Yes Box','Processing...')
    else:
        box.showwarning('No Box','Canceling...')
btn=Button(window,text='Click',command=dialog)
btn.pack(padx=150, pady=50)
window.mainloop()
```

```python
from tkinter import *
import tkinter.messagebox as box
window=Tk()
window.title('Check Button')
frame = Frame(window)

var_1=IntVar()
var_2=IntVar()
var_3=IntVar()

book_1=Checkbutton(frame, text='HTML5', variable=var_1, onvalue=1, offvalue=0)
book_2=Checkbutton(frame, text='CSS3', variable=var_2, onvalue=1, offvalue=0)
book_3=Checkbutton(frame, text='JavaScript', variable=var_3, onvalue=1, offvalue=0)

def dialog():
    str='four Choice'
    if var_1.get()==1: str+='\nHTML5 in easy steps'
    if var_2.get()==1: str+='\nCSS3 in easy steps'
    if var_2.get()==1: str+='\nJavaScript in easy steps'
    box.showinfo('Selection', str)

btn=Button(frame, text='Choose', command=dialog)
btn.pack(side=RIGHT, padx=5)
book_1.pack(side=LEFT)
book_2.pack(side=LEFT)
book_3.pack(side=LEFT)

frame.pack(padx=30,pady=30)

window.mainloop()
```
